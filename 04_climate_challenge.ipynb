## STEP 2: Clean up time series plots by resampling

You may notice that your plot looks a little â€œfuzzyâ€. This happens when
Python is trying to plot a value for every date, but the resolution of
the image is too low to actually do that. You can address this issue by
**resampling** the data, or summarizing it over a time period of your
choice. In this case, we will resample annually, giving us one data
point per year.

<link rel="stylesheet" type="text/css" href="./assets/styles.css"><div class="callout callout-style-default callout-titled callout-task"><div class="callout-header"><div class="callout-icon-container"><i class="callout-icon"></i></div><div class="callout-title-container flex-fill">Try It: Resample</div></div><div class="callout-body-container callout-body"><ol type="1">
<li>Set the frequency of your final data by replacing
<code>DT_OFFSET</code>with a <strong>Datetime Offset Code</strong>.
Check out the table in the <a
href="https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects">pandas
datetime documentation</a> to find the one you want (we recommend the
start of the year).</li>
<li>Choose how to summarize each year of data by replacing
<code>agg_method_here</code> with a method that will calculate the
<strong>average annual value</strong>. Check out the <a
href="https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#basics">pandas
resampling documentation</a> for a list of common built-in options.</li>
<li>Add descriptive comments to the code so the next person reading it
knows what it is doing.</li>
</ol></div></div>

````python
ann_climate_df = (
    climate_df
    .resample('AS')
    .mean()
)
ann_climate_df
````

````python
# Ensure the index is datetime
climate_df.index = pd.to_datetime(climate_df.index)

# Extract the 1990s daily data
climate_1990s = climate_df['1990-01-01':'1999-12-31']

# --- Count valid daily observations per year ---
valid_counts_1990s = climate_1990s['temp_C'].resample('Y').count()

print("Number of valid daily observations per year in the 1990s:")
print(valid_counts_1990s)

# Optional: plot counts interactively
valid_counts_1990s.hvplot(
    title='Valid Daily Observations per Year (1990s)',
    xlabel='Year',
    ylabel='Count',
    width=800,
    height=400,
    marker='[]'
)
# Handle missing daily values
# interpolate missing temperature values
# climate_1990s['temp_C_interp'] = climate_1990s['temp_C'].interpolate(method='time')

# Aggregate annual mean temperatures
# annual_1990s = climate_1990s['temp_C_interp'].resample('Y').mean()

````

````python
# There are only 95 values for 1994, skewing our annual mean


# Slice the period with suspected missing data
missing_period = climate_df['1994-01-01':'1995-12-31']

# Boolean column: True if temp_C is missing
missing_period['is_missing'] = missing_period['temp_C'].isna()

# Convert boolean to integer for plotting
missing_period['is_missing_int'] = missing_period['is_missing'].astype(int)

# Count missing per month (optional)
monthly_missing = missing_period['is_missing_int'].resample('M').sum()
print("Monthly counts of missing daily temperatures (1994â€“1995):")
print(monthly_missing)

# Interactive plot
hvplot_missing = missing_period['is_missing_int'].hvplot(
    title="Missing Daily Temperature Observations (1994â€“1995)",
    ylabel="Missing? (1=missing, 0=present)",
    width=900,
    height=300,
    color='red'
)

hvplot_missing
````
